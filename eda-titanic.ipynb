```python
df_clean=df.drop(["PassengerId"],axis=1,inplace=True)
df_clean=df.drop(["Name"],axis=1)
df_clean=df.drop(["Ticket"],axis=1)
```

# One Hot Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sex'] = le.fit_transform\
                    (df['Sex'])

df['Embarked'] = le.fit_transform\
                    (df['Embarked'])


```

```python
#cleaning Missing values and unnessary columns
df_clean=df
```

```python
df_clean
```

```python
df_clean=df_clean.dropna()
```

```python
df_clean.isnull().sum()
```

```python
#data shape after data cleaning
df_clean.shape
```

```python
df_clean.columns
```

# Finding the Outliers

```python
sns.boxplot(x="Sex",y="Age",data=df_clean)
```

```python
#Shape of Outliers
sns.distplot(df_clean["Age"])
```

```python
#Outliers Removal
df_clean=df_clean[df_clean["Age"]<63]
df_clean.head()
```

# Histogram For All Columns

```python
plt.figure(figsize=(40,10))
df_clean.hist()
```

```python
fig,axes=plt.subplots(2,2,figsize=(20,5))

ax=sns.countplot(x="Survived",data=df_clean,hue="Sex",ax=axes[0,0],palette="cool_r")
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_xticklabels(["Died","Survive"])

ax=sns.countplot(x="Survived",data=df_clean,hue="Pclass",ax=axes[0,1],palette="viridis")
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_xticklabels(["Died","Survive"])


ax=sns.countplot(x="Survived",data=df_clean,hue="SibSp",ax=axes[1,0],palette="cool")
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_xticklabels(["Died","Survive"])


ax=sns.countplot(x="Survived",data=df_clean,hue="Embarked",ax=axes[1,1],palette="cool")
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_xticklabels(["Died","Survive"])
```
